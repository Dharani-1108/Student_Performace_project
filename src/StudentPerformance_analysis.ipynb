```python
#import all the neccessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Satge 3: Data Analysis

#Data Visualization-1

#sum up scores of males in math, read and write respectively
#sum up scores of females in math, read and write respectively
male_scores = clean_data[clean_data['gender'] == 'male'][['math score', 'reading score', 'writing score']].sum()
female_scores = clean_data[clean_data['gender'] == 'female'][['math score', 'reading score', 'writing score']].sum()

# Create pie plots
labels = ['Math', 'Reading', 'Writing']
colors = ['yellow', 'aquamarine','magenta']

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.pie(male_scores, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Male Scores')

plt.subplot(1, 2, 2)
plt.pie(female_scores, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Female Scores')
```

```python
#Data Visualization-2

# Calculate average scores for each parental level of education
avg_scores = clean_data.groupby('parental level of education').mean()

# Create bar plot
plt.figure(figsize=(10, 6))
avg_scores.plot(kind='bar', colormap='viridis')
plt.title('Average Student Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Data Visualization-3

#create count plot for race/ethnicity
sns.countplot(x='race/ethnicity', data=clean_data, color='skyblue')
plt.title('Race/Ethnicity Distribution')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
#Data Visualization-4

# Create a box plot of math scores by race/ethnicity
plt.figure(figsize=(10, 6))
sns.boxplot(data=clean_data, x='race/ethnicity', y='math score', color='lightcoral')
plt.title('Math Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Math Score')
plt.show()
```

```python
#Data Visualization-5

# Create joined scatter plot for reading, writing and math scores
plt.figure(figsize=(8, 6))
sns.scatterplot(data=clean_data, x='reading score', y='writing score', hue='math score', palette='viridis')
plt.title('Joined Scatter Plot of Reading, Writing, and Math Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.legend(title='Math Score')
plt.grid(True)
plt.show()
```
